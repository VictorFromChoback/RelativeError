```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
import seaborn as sns

import typing as tp
from tqdm import tqdm_notebook
```

```python
ALPHA = 3

def calc_next(cur, h):
    dif = -cur * h + ALPHA * np.abs(cur) * sps.norm.rvs(loc=0, scale=np.sqrt(h))
    return cur + dif

def gen_process(start, iters = 100, h = 1):
    data = [start]
    for j in range(iters):
        data.append(calc_next(data[-1], h))
    return data
    
```

```python
with sns.plotting_context("notebook"), sns.axes_style("darkgrid"):
    plt.figure(figsize=(10, 5))
    iters = 30
    start = 10
    h = 0.1
    for k in range(10):
        plt.plot(np.arange(iters + 1), gen_process(start, iters, h), alpha=0.3)
    plt.show()
```
